2. The tutorial mentions two other kernels: polynomial and rbf.  Modify the model so that it uses the rbf kernel with 
gamma = 2. For documentation on gamma, see https://scikit-
learn.org/stable/modules/generated/sklearn.svm.SVC.html 

```python
from sklearn import datasets #Import scikit-learn dataset
from sklearn.model_selection import train_test_split# Train test split
from sklearn import svm #Import SVM
from sklearn import metrics #import accuracy
import numpy as np # linear algebra

cancer = datasets.load_breast_cancer() #load datasets
 # Split the data 80 percent train and 20 percent test
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, test_size = 0.2, random_state = 382)
 
 # Create the classifier
clf = svm.SVC(kernel = 'rbf', gamma = 2)
clf2 = svm.SVC(kernel = 'linear')
# Train the model
clf.fit(X_train, y_train)
clf2.fit(X_train, y_train)
 # Predict the response for test dataset
y_pred = clf.predict(X_test)
y_pred2 = clf2.predict(X_test)
 # Model Accuracy
print('The accuracy is:{}'.format(metrics.accuracy_score(y_test, y_pred)))
print('The accuracy linear is:{}'.format(metrics.accuracy_score(y_test, y_pred2)))
  # Model precision
print('The precision is:{}'.format(metrics.precision_score(y_test, y_pred)))
print('The precision linear is:{}'.format(metrics.precision_score(y_test, y_pred2)))
  # Model recall
print('The recall is:{}'.format(metrics.recall_score(y_test, y_pred)))
print('The recall linear is:{}'.format(metrics.recall_score(y_test, y_pred2)))

```

4. Investigate the effect of training/testing split on model accuracy, precision and recall for the data in 3.  You may do 
this by making a table, where each row of the table corresponds to a training/testing split, and the columns of the 
table are: training proportion, accuracy, recall, precision. Display your results by graphing accuracy, recall, and 
precision as a function of training proportion. Use a loop to generate results for multiple training/testing splits. 
You should try about 25 splits, with training proportion between 0.04 and 0.96

```python
import numpy as np # linear algebra
from sklearn.model_selection import train_test_split# Train test split
from sklearn import svm #Import SVM
from sklearn import metrics #import accuracy
import matplotlib.pyplot as plt #for graphs

#Generate random data
data = np.random.uniform(-1,1,(200,3))
# classify
tc = data[:,0]*data[:,1]*data[:,2]>0
# Convert to binary
test_class = tc*1

# Set up an array for different splits:
split = np.linspace(0.1,0.9,1000)

#Set Up arrays to store accuracy, precision and recall
accuracy_array = np.empty(len(split))
precision_array = np.empty(len(split))
recall_array = np.empty(len(split))

 # loop over splits
for i in range(len(split)):
   X_train, X_test, y_train, y_test = train_test_split(data, test_class, test_size = split[i])
# Create a SVM classifier with linear kernel
   clf = svm.SVC(kernel = 'linear')
# Train the model
   clf.fit(X_train, y_train)

 # Predict the response for test dataset
   y_pred = clf.predict(X_test) 
  # Model Accuracy
   accuracy_array[i] = metrics.accuracy_score(y_test, y_pred)
  # Model precision
   precision_array[i] = metrics.precision_score(y_test, y_pred)
  # Model recall
   recall_array[i] = metrics.recall_score(y_test, y_pred)

# Generate plot
plt.title('SVM: Varying Test Size')
plt.plot(split, accuracy_array,'.', label = 'Accuracy')
plt.plot(split, precision_array,'.', label = 'Precision')
plt.plot(split, recall_array,'.', label = 'Recall')
plt.legend()
plt.xlabel('Number of Splits')
plt.ylabel('Scores')
plt.show()
```

We may choose split as 0.4 where the alll these three metrics are higher than the majority of the graph.

5. Using your data from 3. above, find the best kernel and kernel parameters. Compare polynomial, linear, and rbf kernels.  
Note that the rbf kernel requires a ‘gamma’ parameter: you may search for values of gamma in the interval [0.5, 4], and 
search for polynomial degrees between 1 and 5. 

```python
# Set up kernels:
kernels = ['linear', 'poly', 'rbf']

#Set Up arrays to store accuracy, precision and recall
accuracy_array = np.empty(len(kernels))
precision_array = np.empty(len(kernels))
recall_array = np.empty(len(kernels))

# loop over kernels

X_train, X_test, y_train, y_test = train_test_split(data, test_class, test_size = 0.4)
# Create a SVM classifier with linear kernel
for i in range(len(kernels)):
   clf = svm.SVC(kernel = kernels[i])
# Train the model
   clf.fit(X_train, y_train)

 # Predict the response for test dataset
   y_pred = clf.predict(X_test) 
  # Model Accuracy
   accuracy_array[i] = metrics.accuracy_score(y_test, y_pred)
  # Model precision
   precision_array[i] = metrics.precision_score(y_test, y_pred)
  # Model recall
   recall_array[i] = metrics.recall_score(y_test, y_pred)

# Generate plot
plt.title('SVM: Varying Kernels')
plt.plot(kernels, accuracy_array, label = 'Accuracy')
plt.plot(kernels, precision_array, label = 'Precision')
plt.plot(kernels, recall_array, label = 'Recall')
plt.legend()
plt.xlabel('Kernels')
plt.ylabel('Scores')
plt.show()
```

It looks like the highest scores are obtained by rbf
